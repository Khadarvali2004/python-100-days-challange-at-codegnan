#type of arguments
1.positional arguments
2.keyword arguments
3.default arguments
4.varible length positional argument
5.varible length keyword argument


```python
#1.positional arguments
def sub(x,y):
    return x -y
b1=int(input())
b2=int(input())
print(sub(b1,b2))
print(sub(b2,b1))
```

```python
#2.keyword arguments
def sub(x,y):
    return x -y
b1=int(input())
b2=int(input())
print(sub(x=b1,y=b2))
print(sub(y=b2,x=b1))
```
