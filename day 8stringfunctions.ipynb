```python
#string functions
#case converstions
s="   'my' name is khadar vali     "
print(s.upper())
print(s.lower())
print(s.swapcase())
print(s.title())
print(s.capitalize())
#sstring manuplication function
print(s.split())
print(s.replace('is','was'))
print(s.lstrip())
print(s.rstrip())
print(s.join(['i','am','vali']))
print(s.index('is'))
```
