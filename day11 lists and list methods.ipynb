```python
#list methods
#adding elements to list
#append(ele):adds the elements at end of the list
#extend(list):adds all elements in list at end
#insert(pop,ele):insert
l=[1,2,3,4,]
l.append(20)
print(l)
l.extend([10,20,30,40,52,28])
print(l)
l.insert(3,250)
print(l)
l.insert(-1,100)
print(l)
```
