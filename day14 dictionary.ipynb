```python
#dictionary values can access by using keys
details={'name':"khadar","batch":49,"course":"python",'name':'khadar vali'}
print(details['name'])
```

```python
#dictionary keys are unique
details={'name':"khadar","batch":49,"course":"python",'name':'khadar vali'}
print(details[0])
print(details[1])
```

```python
#update dictionary value ,#dictionary is mutable
details={'name':"khadar","batch":49,"course":"python",'name':'khadar vali'}
details['course']='java'
details['age']=50
details['batch']=47
print(details)

```

```python
#dictionary methods
#get(key,default_value):It returns the value if the key is present in the dictionary
#otherwise it returns default value
details={'name':"khadar","batch":49,"course":"python",'name':'khadar vali'}
print(details.get('location','india'))
print(details.get('name','ramu'))
print(details.get('marks',20))
print(details.get('score'))
```

```python
#update(items):dictionary updates with new items
details={'name':"khadar","batch":49,"course":"python",'name':'vali'}
new_dict={'course':'java','batch':32,'location':'india'}
details.update(new_dict)
print(details)
```

```python
#keys():It returns the list of dictionary keys
#values():It returns the list of dictionary values
#items():it returns the dictionary lisst of key-values pairs
details={'name':"khadar","batch":49,"course":"python",'name':'vali'}
print(details.keys())
print(details.values())
print(details.items())
```

```python
#pop(key):It returns the values and remove the items from dictionary if key is present,otherwise keyerror
#popitem():It returns and removes last item
details={'name':"khadar","batch":49,"course":"python",'name':'vali'}
print(details.pop('name'))
print(details.popitem())
print(details)

```
